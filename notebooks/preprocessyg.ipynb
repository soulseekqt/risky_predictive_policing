4. Splitting date column into months. (January to December)

```python
import pandas as pd
```

```python
data = pd.read_csv('raw_data/chicago.csv')

data.head()
```

```python
print(data.columns)
```

```python

data['DATE__OF_OCCURRENCE'] = pd.to_datetime(data['DATE__OF_OCCURRENCE'])

# Extraction of the month name and creation a new column
data['Month'] = data['DATE__OF_OCCURRENCE'].dt.month_name()

# Display of the first 50 rows
data[['DATE__OF_OCCURRENCE', 'Month']].head(50)
```

```python
# last 50 rows
data[['DATE__OF_OCCURRENCE', 'Month']].tail(50)
```
